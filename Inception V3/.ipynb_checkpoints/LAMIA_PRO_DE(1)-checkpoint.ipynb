```python
# Define the consistent order
consistent_classes = ['Adenocarcinoma', 'Large cell carcinoma','Normal', 'Squamous cell carcinoma']

# Reorder test_classes to match the consistent order
test_classes = [class_name for class_name in consistent_classes if class_name in test_classes]

print(test_classes)
print(train_classes)
print(vali_classes)
```

# **Applying data augmentation**

Applying data augmentation, which applies changes on the training

```python
train = ImageDataGenerator(

                           rotation_range =10,
                           width_shift_range =0.2,
                           height_shift_range =0.2,
                           shear_range =0.2,
                           zoom_range =0.2,
                           horizontal_flip = True,
                           vertical_flip = False,
                           dtype = 'float32', #optimization reason
                           preprocessing_function=tf.keras.applications.inception_v3.preprocess_input
                           )
vali = ImageDataGenerator(

                          dtype = 'float32', #optimization reason
                          preprocessing_function=tf.keras.applications.inception_v3.preprocess_input
                          )
test = ImageDataGenerator(

                          dtype = 'float32', #optimization reason
                          preprocessing_function=tf.keras.applications.inception_v3.preprocess_input

                           )
```

# **Spliting data into training/validation/testing**

```python

training_data = train.flow_from_directory(train_path,
                                          target_size =(224,224),
                                          batch_size = 64,
                                          class_mode ='categorical')

validation_data = vali.flow_from_directory(validation_path,
                                           target_size=(224,224),
                                           batch_size =64,
                                           class_mode = 'categorical')

testing_data = test.flow_from_directory(testing_path,
                                        target_size =(224,224),
                                        batch_size =64,
                                        class_mode = 'categorical',
                                        shuffle = False,

                                        )
```

# **Counting and Showing**

This section contains plots showing how many data from each class I have

```python
#A function that counts the quantity of each class
def counting_class(generator):
  labels = generator.labels
  class_counts = np.bincount(labels)
  return class_counts

train_class_counts = counting_class(training_data)
vali_class_counts = counting_class(validation_data)
test_class_counts = counting_class(testing_data)

class_labels = list(training_data.class_indices.keys())

#A function that plots the quantity of each class
def plotting_histogram(class_counts, title, class_labels):
  plt.figure(figsize=(10,4))
  plt.bar(range(len(class_counts)), class_counts, tick_label = class_labels)
  plt.title(title)
  plt.xlabel('Class')
  plt.ylabel('Quantity')
  plt.show
```

```python
print(train_class_counts)
print(vali_class_counts)
print(test_class_counts)
```

```python
plotting_histogram(train_class_counts, 'Training Data Distribution', class_labels)
```

```python
plotting_histogram(vali_class_counts, 'Validation Data Distribution', class_labels)
```

```python
plotting_histogram(test_class_counts, 'Testing Data Distribution', class_labels)
```

# **Neural Network**

```python
from keras.callbacks import ModelCheckpoint
from keras.applications import InceptionV3
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
inceptionv3 = InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
inceptionv3.trainable = False
```

```python
model = Sequential()

model.add(inceptionv3)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(4, activation='softmax'))
```

# **Compiling the model**

```python
from tensorflow.keras.callbacks import EarlyStopping


model.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics= ['categorical_accuracy'])

earlystopper = EarlyStopping(patience = 10, verbose = 1)
```

# **K-fold cross validation**

# **training the model**

```python
epochs = 200
history = model.fit(
    training_data,
    validation_data=validation_data,
    epochs=epochs,
    callbacks = earlystopper,
    verbose=1
)
```

# **Saving the model**

```python
#Saving the model's architecture
json_model = model.to_json()
with open("/content/gdrive/MyDrive/My_Model's_score/Model's Architecture" + 'model_architecture_DE.json', 'w') as json_file:
  json_file.write(json_model)

#Saving the model's weights
model.save_weights("/content/gdrive/MyDrive/My_Model's_score/Model's Weights/" + 'model_weights_DE.h5')
```

# **Model's accuracy and loss during training**

```python
def plot_history_model(history):
  fig,axs = plt.subplots(1,2,figsize = (15,5))
  axs[0].plot(range(1, len(history.history['categorical_accuracy']) +1 ), history.history['categorical_accuracy'], 'r')
  axs[0].plot(range(1,len(history.history['val_categorical_accuracy']) + 1), history.history['val_categorical_accuracy'], 'b')
  axs[0].set_title("Model's accuracy")
  axs[0].set_ylabel('Accuracy')
  axs[0].set_xlabel('Epoch')
  axs[0].legend(['training accuracy', 'validation accuracy'], loc = 'best')

  axs[1].plot(range(1, len(history.history['loss']) +1 ), history.history['loss'], 'r')
  axs[1].plot(range(1,len(history.history['val_loss']) + 1), history.history['val_loss'], 'b')
  axs[1].set_title("Model's loss")
  axs[1].set_ylabel('Loss')
  axs[1].set_xlabel('Epoch')
  axs[1].legend(['training loss', 'validation loss'], loc = 'best')

plot_history_model(history)
```

# **Model's accuracy**

```python
model_score = model.evaluate(testing_data)
```

```python
import seaborn as sns

predictions_proba = model.predict(testing_data)
predictions = np.argmax(predictions_proba, axis =1)
true_label = testing_data.classes
reporting = classification_report(true_label ,predictions)
print(reporting)
```

```python
confusion_matrix = confusion_matrix(true_label, predictions)
sns.heatmap(confusion_matrix, fmt ='g', annot = True, cmap = 'Reds', xticklabels=test_classes,
            yticklabels = test_classes,)

plt.xlabel("Predictions")
plt.ylabel("True label")
plt.title("Confusion Matrix")
plt.xticks(rotation = 45)
plt.show()
```
