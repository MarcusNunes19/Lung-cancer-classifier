# **Applying data augmentation**

Applying data augmentation, which applies changes on the training

```python
train = ImageDataGenerator(

                           rotation_range =10,
                           width_shift_range =0.2,
                           height_shift_range =0.2,
                           shear_range =0.2,
                           zoom_range =0.2,
                           horizontal_flip = True,
                           vertical_flip = False,
                           dtype = 'float32', #optimization reason
                           preprocessing_function=tf.keras.applications.inception_v3.preprocess_input
                           )
vali = ImageDataGenerator(

                          dtype = 'float32', #optimization reason
                          preprocessing_function=tf.keras.applications.inception_v3.preprocess_input
                          )
test = ImageDataGenerator(

                          dtype = 'float32', #optimization reason
                          preprocessing_function=tf.keras.applications.inception_v3.preprocess_input

                           )
```

# **Spliting data into training/validation/testing**

```python

training_data = train.flow_from_directory(train_path,
                                          target_size =(224,224),
                                          batch_size = 64,
                                          class_mode ='categorical')

validation_data = vali.flow_from_directory(validation_path,
                                           target_size=(224,224),
                                           batch_size =64,
                                           class_mode = 'categorical')

testing_data = test.flow_from_directory(testing_path,
                                        target_size =(224,224),
                                        batch_size =64,
                                        class_mode = 'categorical',
                                        shuffle = False,

                                        )
```

# **Counting and Showing**

This section contains plots showing how many data from each class I have

```python
#A function that counts the quantity of each class
def counting_class(generator):
  labels = generator.labels
  class_counts = np.bincount(labels)
  return class_counts

train_class_counts = counting_class(training_data)
vali_class_counts = counting_class(validation_data)
test_class_counts = counting_class(testing_data)

class_labels = list(training_data.class_indices.keys())

#A function that plots the quantity of each class
def plotting_histogram(class_counts, title, class_labels):
  plt.figure(figsize=(10,4))
  plt.bar(range(len(class_counts)), class_counts, tick_label = class_labels)
  plt.title(title)
  plt.xlabel('Class')
  plt.ylabel('Quantity')
  plt.show
```

```python
print(train_class_counts)
print(vali_class_counts)
print(test_class_counts)
```

```python
plotting_histogram(train_class_counts, 'Training Data Distribution', class_labels)
```

```python
plotting_histogram(vali_class_counts, 'Validation Data Distribution', class_labels)
```

```python
plotting_histogram(test_class_counts, 'Testing Data Distribution', class_labels)
```

# **Neural Network**

```python
from keras.callbacks import ModelCheckpoint
from keras.applications import InceptionV3
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
inceptionv3 = InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
inceptionv3.trainable = False
```

```python
model = Sequential()

model.add(inceptionv3)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(4, activation='softmax'))
```

# **Compiling the model**

```python
from tensorflow.keras.callbacks import EarlyStopping


model.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics= ['categorical_accuracy'])

earlystopper = EarlyStopping(patience = 10, verbose = 1)
```

# **K-fold cross validation**

```python
from sklearn.model_selection import KFold
```

## **Extracting paths and labels**

```python
def get_paths_labels(generator):
  img_paths = []
  y_labels = []
  classes_labels = {v: k for k, v in generator.class_indices.items()}

  for class_directory, index in generator.class_indices.items():
    class_path = os.path.join(generator.directory, class_directory)
    if os.path.isdir(class_path):
      for img in os.listdir(class_path):
        img_path = os.path.join(class_path, img)
        img_paths.append(img_path)
        y_labels.append(index)

  return np.array(img_paths), np.array(y_labels)
```

```python
all_paths, all_labels = get_paths_labels(training_data)
all_labels = np.array([str(label) for label in all_labels])

kf = KFold(n_splits=5, shuffle = True, random_state= 42)
results = []
```

```python
for train_index, val_index in kf.split(all_paths):
  train_img_paths, val_img_paths = all_paths[train_index], all_paths[val_index]
  train_labels, val_labels = all_labels[train_index], all_labels[val_index]
```

## **Data generators for each fold**

```python
train_gen = train.flow_from_dataframe(
        dataframe=pd.DataFrame({'filename': train_img_paths, 'class': train_labels}),
        directory=None,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=64,
        class_mode='categorical',
        dtype = 'float32',
        preprocessing_function=tf.keras.applications.inception_v3.preprocess_input

    )

val_gen = vali.flow_from_dataframe(
        dataframe=pd.DataFrame({'filename': val_img_paths, 'class': val_labels}),
        directory=None,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=64,
        dtype = 'float32',
        class_mode='categorical',
        preprocessing_function=tf.keras.applications.inception_v3.preprocess_input
    )
```

# **training the model**

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopper = EarlyStopping(patience = 10, restore_best_weights = True)

model.fit(train_gen,
          validation_data =val_gen,
          epochs = 100,
          callbacks = [early_stopper])
```

# **Model's accuracy**

```python
validation_loss, validation_accuracy = model.evaluate(val_gen)
```

```python
print(f"Fold {len(results) + 1}: Validation Loss = {validation_loss}, Validation Accuracy = {validation_accuracy}")
results.append((validation_loss, validation_accuracy))
```

## **Average results**

```python
average_loss = np.mean([result[0] for result in results])
average_accuracy = np.mean([result[1] for result in results])
print(f"The average validation loss is {average_loss}")
print(f"The average accuracy is {average_accuracy}")
```
